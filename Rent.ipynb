```python
#8

amenities_columns = ['gym', 'lift', 'swimming_pool']

# Create an empty correlation matrix
correlation_matrix = np.zeros((len(amenities_columns), 2))

# Calculate Point-Biserial Correlation Coefficient for each amenity
for i, amenity in enumerate(amenities_columns):
    correlation_coefficient, _ = pointbiserialr(df2_cleaned[amenity], df2_cleaned['rent'])
    correlation_matrix[i, 0] = correlation_coefficient
    correlation_matrix[i, 1] = abs(correlation_coefficient)  # Store absolute values for ranking

# Create a DataFrame for better visualization
correlation_df = pd.DataFrame(correlation_matrix, columns=['Correlation', 'Absolute_Correlation'], index=amenities_columns)

# Print the correlation matrix
print("Point-Biserial Correlation Coefficient Matrix: Rent and Amenities")
print(correlation_df)
```
